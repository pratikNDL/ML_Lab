```python
import sklearn
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.describe()
```

```python
x = df.drop('price', axis = 'columns')
x
```

```python
y = df.price
y
```

```python
model = linear_model.LinearRegression()
model.fit(x , y)
```

```python
model.predict([[3350]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
3350*136.39361233 + 150543.37795405491
```

```python
p = model.predict(area_df)
p
```

```python
area_df['y']=p
area_df
```

```python
area_df.to_csv("prediction.csv")
```

```python
%matplotlib inLine
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(df.area, df.price, color ='blue', marker = '+')
plt.plot(df.area, model.predict(df[['area']]), color='red')
```
